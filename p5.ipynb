```python
import pandas as pd
import matplotlib.pyplot as plt

# Load cleaned dataset
df = pd.read_excel('cleaned_hotels.xlsx')

# Ensure correct column names
if 'PPN' not in df.columns or 'Beds' not in df.columns:
    raise KeyError("Missing required columns: 'PPN' or 'Beds'")

# Convert columns to numeric values (if needed)
df['PPN'] = pd.to_numeric(df['PPN'], errors='coerce')
df['Beds'] = pd.to_numeric(df['Beds'], errors='coerce')

# Drop rows with missing values
df = df.dropna(subset=['PPN', 'Beds'])

# Plot graph
plt.scatter(df['Beds'], df['PPN'], color='blue', alpha=0.5)
plt.xlabel('Number of Bedrooms')
plt.ylabel('Price Per Night (PPN)')
plt.title('PPN vs Number of Bedrooms')
plt.show()
```
