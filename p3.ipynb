```python
import pandas as pd

# Load the cleaned dataset
df = pd.read_excel('cleaned_hotels.xlsx')

# Ensure the 'Reviews' column exists and is numeric
df['Reviews'] = pd.to_numeric(df['Reviews'], errors='coerce')

# Check if 'PType' and 'Reviews' exist
if 'PType' not in df.columns:
    raise KeyError("Column 'PType' not found in the dataset!")

if 'Reviews' not in df.columns:
    raise KeyError("Column 'Reviews' not found in the dataset!")

# Get the best room (highest Reviews) under each Type (PType) in each City
best_rooms = df.loc[df.groupby(['City', 'PType'])['Reviews'].idxmax().dropna()]

# Save the result to an Excel file
best_rooms.to_excel('best_rooms_per_type_per_city.xlsx', index=False)

# Print the result
print(best_rooms)
```
